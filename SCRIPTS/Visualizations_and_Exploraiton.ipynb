```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
# prompt: create a bar graph with counts of each Ripening Index Classification and color code

# Count occurrences of each ripening index classification
ripening_counts = df['Ripening Index Classification'].value_counts()

# Create a bar graph with color coding
plt.figure(figsize=(10, 6))
sns.barplot(x=ripening_counts.index, y=ripening_counts.values, palette='viridis')
plt.xlabel('Ripening Index Classification')
plt.ylabel('Count')
plt.title('Counts of Ripening Index Classifications')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
import os

# Mount the drive
drive.mount('/content/drive')

# Specify the dataset path
dataset_path = '/content/drive/MyDrive/DS Project 3/Avocado Ripening Dataset'

# List files to confirm access
print(os.listdir(dataset_path))

# Function to calculate brightness
def calculate_brightness(image_path):
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    brightness = np.mean(np.array(img))
    return brightness

# Iterate through all images and calculate brightness
brightness_values = []
image_files = [f for f in os.listdir(dataset_path) if f.endswith('.jpg')]  # Get all .jpg files
for image_file in image_files:
    image_path = os.path.join(dataset_path, image_file)  # Construct the full path
    brightness_values.append(calculate_brightness(image_path))

# Plot the brightness distribution
plt.hist(brightness_values, bins=50, color='blue', alpha=0.7)
plt.title('Brightness Distribution')
plt.xlabel('Brightness')
plt.ylabel('Frequency')
plt.show()
```

```python
# Function to get image dimensions
def get_image_dimensions(image_path):
    img = Image.open(image_path)
    return img.size  # Returns (width, height)

# Iterate through all images and collect dimensions
dimensions = []
for image_file in image_files:
    image_path = os.path.join(dataset_path, image_file)
    dimensions.append(get_image_dimensions(image_path))

# Split dimensions into width and height
widths, heights = zip(*dimensions)

# Plot distributions of width and height
plt.hist(widths, bins=20, alpha=0.7, label='Width')
plt.hist(heights, bins=20, alpha=0.7, label='Height')
plt.title('Image Dimensions Distribution')
plt.xlabel('Pixels')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Function to calculate average RGB values
def calculate_avg_rgb(image_path):
    img = Image.open(image_path).convert('RGB')
    img_array = np.array(img)
    r_mean = np.mean(img_array[:, :, 0])
    g_mean = np.mean(img_array[:, :, 1])
    b_mean = np.mean(img_array[:, :,2])
    return r_mean, g_mean, b_mean

# Collect RGB averages for all images
r_values, g_values, b_values = [], [], []
for image_file in image_files:
    image_path = os.path.join(dataset_path, image_file)
    r, g, b = calculate_avg_rgb(image_path)
    r_values.append(r)
    g_values.append(g)
    b_values.append(b)

# Plot distributions of RGB channels
plt.hist(r_values, bins=50, alpha=0.7, color='red', label='Red Channel')
plt.hist(g_values, bins=50, alpha=0.7, color='green', label='Green Channel')
plt.hist(b_values, bins=50, alpha=0.7, color='blue', label='Blue Channel')
plt.title('Color Intensity Distribution')
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
